```python
#Top 10 States by Sales and Their Percentage
state_sales=sales.groupby('State')['Sales'].sum()
state_sales_percent=round((state_sales/state_sales.sum())*100,2)
result=pd.concat([state_sales,state_sales_percent],axis=1)
result.columns=['Sales','Sales %']
result.sort_values('Sales', ascending=False).head(10)
```

```python
#Category-wise Sales and Their Percentage of Total Sales
category_sales=sales.groupby('Category')['Sales'].sum()
category_sales_percent=round((category_sales/category_sales.sum())*100,2)
result=pd.concat([category_sales,category_sales_percent],axis=1)
result.columns=['Sales','Sales %']
result.sort_values('Sales', ascending=False)
```

```python
#Top 10 sub-categories
sales['Sub-Category'].value_counts().head(10)
```

```python
#Sales by Sub-Categories and Their Percentage of Total Sales
sub_category_sales=sales.groupby('Sub-Category')['Sales'].sum()
sub_category_sales_percent=round((sub_category_sales/sub_category_sales.sum())*100,2)
result=pd.concat([sub_category_sales,sub_category_sales_percent],axis=1)
result.columns=['Sales','Sales %']
result.sort_values('Sales',ascending=False).head(10)
```

```python
#Sales by Region
region_sales=sales.groupby('Region')['Sales'].sum()
region_sales_percent=round((region_sales/region_sales.sum())*100,2)
result=pd.concat([region_sales,region_sales_percent],axis=1)
result.columns=['Sales','Sales %']
result.sort_values('Sales', ascending=False)
```

```python
#new code: lifetime value(LTV)
sales['Order Date']=pd.to_datetime(sales['Order Date'], dayfirst=True,errors='coerce')
ltv=sales.groupby('Customer Name').agg({'Sales':'sum', 'Order Date':[min,max,'nunique']})
ltv.columns=['Total Sales', 'First Order', 'Last Order', 'Number of Orders']
ltv['Costumer Lifetime days']=(ltv['Last Order']-ltv['First Order']).dt.days+1
ltv['Avg Sales per day']=ltv['Total Sales']/ltv['Costumer Lifetime days']
ltv=ltv.sort_values('Total Sales', ascending=False).head(10)
ltv
```

```python
ltv.head(10).plot(kind='bar', figsize=(10,5),title='Top 10 Customers by Sales')
plt.show()
```

```python
#new code : regional trend by year
region_year= sales.groupby(['Region','Year'])['Sales'].sum().sort_values(ascending=False).unstack()
region_year.plot(kind='bar')
plt.title('Regional Trend by Year')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
#line plot
```

```python
#Sales by Year
sales['Year']=sales['Order Date'].dt.year
sales_by_year= sales.groupby('Year')['Sales'].sum()*100/sales['Sales'].sum()
sales_by_year.round(2)
```

```python
#new code : Sales by season
import matplotlib.pyplot as plt
sales['Month']=sales['Order Date'].dt.month
def get_season(month):
    if month in [3,4,5]:
        return 'Spring'
    elif month in [6,7,8]:
        return 'Summer'
    elif month in [9,10,11]:
        return 'Fall'
    else:
        return 'Winter'

sales['Season']=sales['Month'].apply(get_season)
seasonal_sales=sales.groupby(['Season','Year'])['Sales'].sum().unstack()
seasonal_sales.plot(kind='bar', title='Seasonal Sales')
plt.xlabel('Season')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
#line plot y=sales
```

```python
#Average shipping day
sales['Order Date']=pd.to_datetime(sales['Order Date'], dayfirst=True,errors='coerce')
sales['Ship Date']=pd.to_datetime(sales['Ship Date'], dayfirst=True,errors='coerce')
sales['Shipping Days']=(sales['Ship Date']-sales['Order Date']).dt.days
round(sales['Shipping Days'].mean())
```

```python
#Sales by month
monthly_sales=sales.groupby('Month')['Sales'].sum().sort_index()
plt.figure(figsize=(10,5))
monthly_sales.plot(kind='bar')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Top 10 States by Sales
top_states=sales.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
top_states.plot(kind='bar', color='orange')
plt.title('Top 10 States by Sales')
plt.xlabel('State')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Top 10 customers by Sales
top_customers=sales.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
top_customers.plot(kind='bar', color='green')
plt.title('Top 10 Customers by Sales')
plt.xlabel('Customer Name')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Sales by category
category_sales=sales.groupby('Category')['Sales'].sum()
plt.figure(figsize=(6,4))
category_sales.plot(kind='bar', color='purple')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#Top 10 Sub-categories by Sales
top_subcategories=sales.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
top_subcategories.plot(kind='bar', color='red')
plt.title('Top 10 Sub-Categories by Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Top 10 cities by Sales
top_cities=sales.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
top_cities.plot(kind='bar', color='blue')
plt.title('Top 10 Cities by Sales')
plt.xlabel('City')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Sales by Ship Mode
ship_mode_sales=sales['Ship Mode'].value_counts()
plt.figure(figsize=(6,4))
ship_mode_sales.plot(kind='bar',color='skyblue')
plt.title('Sales by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#Sales by Region
region_sales =sales.groupby('Region')['Sales'].sum().sort_values()
plt.figure(figsize=(6,4))
region_sales.plot(kind='bar', color='orange')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#Distribution of Segments
segment_counts=sales['Segment'].value_counts()
plt.figure(figsize=(6,4))
segment_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightyellow'])
plt.title('Distribution of Segments')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

#Conclusion

The analysis revealed several key insights:
- The 'Consumer' segment generated the highest total sales.
- 'Standard Class' was the most frequently used shipping method.
- The 'West' region had the largest number of orders.
- Product categories such as 'Chairs' and 'Phones' significantly contributed to total revenue.
- Sales trends showed clear monthly and seasonal fluctuations.
- The 'Top 10 customers' accounted for the notable share of overall sales.
- New York and Los Angeles are the top cities in terms of total sales.
- Technology is the most sold product category, followed by Furniture and Office Supplies.

These findings may support data-driven decisions in marketing, logistics, and inventory planning.
