#US Births by Year, State, and Education Level

This dataset provides birth rates and related data across the 50 states and DC from 2016 to 2021.   The data  includes detailed information such as number of births, gender, birth weight, state, and year of the delivery. A particular emphasis is given to detailed information on the mother's educational level. With this dataset, one can, for example, examine trends and patterns in birth rates across different academic groups and geographic locations.

link to source: https://www.kaggle.com/datasets/danbraswell/temporary-us-births

##Description:
- State – Full name of the U.S. state where the birth was recorded
- State Abbreviation – Two-letter postal abbreviation for the state
- Year – The year when the births were recorded
- Gender – The sex of the newborn
- Education Level of Mother – The descriptive education level of the mother at the time of birth
- Education Level Code – Numeric code representing the mother’s education level, usually mapped to categories
- Number of Births – Total count of births recorded for the specified grouping
- Average Age of Mother (years) – Average age in years of mothers for the births in that group
- Average Birth Weight (g) – Average weight of the newborns in grams for the given group

```python
#Number of Births by Gender
grouped_gender=df.groupby('Gender')['Number of Births'].sum()
grouped_gender_percent=((grouped_gender/grouped_gender.sum()*100)).round(1).astype(str)+'%'
result=pd.concat([grouped_gender,grouped_gender_percent],axis=1)
result.columns=['Number of Births', 'Percentage']
result
```

```python
#Distribution of Births by Gender with Percent Labels
grouped_gender_percent=((grouped_gender/grouped_gender.sum()*100)).round(1)
plt.figure(figsize=(6,4))
plt.pie(
    grouped_gender,
    labels=[f'{gender}' for gender, percent in zip(grouped_gender.index, grouped_gender_percent)],
    autopct='%1.1f%%',
    colors=['lightpink', 'lightblue'],
    startangle=90
)
plt.title('Number of Births by Gender')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
#Average Newborn Weight by Sex
df.groupby('Gender')['Average Birth Weight (g)'].mean().round()
```

```python
#Relation Between Mother’s Age and Birth Weight
plt.figure(figsize=(8, 6))
sns.regplot(x='Average Age of Mother (years)', y='Average Birth Weight (g)', data=df)
plt.title('Relation Between Mother’s Age and Birth Weight')
plt.xlabel('Average Age of Mother (years)')
plt.ylabel('Average Birth Weight (g)')
plt.tight_layout()
plt.show()
```

```python
#Analysis of Number of Births by Mother's Age Group

df['Rounded Age'] = df['Average Age of Mother (years)'].round()

bins = list(range(23, 38, 2))
labels = [f'{i}-{i+1}' for i in bins[:-1]]


df['Age Group'] = pd.cut(df['Rounded Age'], bins=bins, labels=labels, right=False)


grouped_births = df.groupby('Age Group')['Number of Births'].sum()
grouped_births_percent = round(grouped_births / grouped_births.sum() * 100, 2).astype(str) + '%'

result = pd.concat([grouped_births, grouped_births_percent], axis=1)
result.columns = ['Number of Births', 'Percentage']
result.sort_values('Number of Births', ascending=False)
```

```python
#Number of Births by Mother's Age Group
grouped_births_percent=round(grouped_births/grouped_births.sum()*100,2)
plt.figure(figsize=(6,5))
bars = plt.bar(grouped_births_percent.index, grouped_births_percent.values)
plt.title('Number of Births by Mother\'s Age Group')
plt.xlabel('Age Group')
plt.ylabel('Percentage')
plt.tight_layout()

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.1, f'{height}%',
             ha='center', va='bottom')

plt.show()
```

```python
#Education Level Distribution Across Age Groups
sns.barplot(data=df,x='Age Group',y='Education Level Code', ci=None)
plt.show()
```

```python
#Average Age of Mother by Education Level
df.groupby('Education Level of Mother')['Average Age of Mother (years)'].mean().round()
```

```python
#Births summed by education level
edu_map = {
    1: '8th grade or less',
    2: '9th through 12th grade with no diploma',
    3: 'High school graduate or GED completed',
    4: 'Some college credit, but not a degree',
    5: 'Associate degree (AA, AS)',
    6: "Bachelor's degree (BA, AB, BS)",
    7: "Master's degree (MA, MS, MEng, MEd, MSW, MBA)",
    8: 'Doctorate (PhD, EdD) or Professional Degree (MD, DDS, DVM, LLB, JD)',
    -9: 'Unknown or Not Stated'
}
df['Education Level'] = df['Education Level Code'].map(edu_map)
births_by_education = df.groupby(['Education Level','Education Level Code'])['Number of Births'].sum()
births_by_education_percent = (births_by_education / births_by_education.sum() * 100).round(1).astype(str) + '%'
result=pd.concat([births_by_education,births_by_education_percent],axis=1)
result.columns=['Number of Births','Percentage']
result = result.reset_index().sort_values('Education Level Code')
result
```

```python
#Percentage of Births by Education Level Code
import matplotlib.patches as mpatches
def get_color(code):
    if code in [1, 2, 3]:
        return 'red'
    elif code in [4, 5]:
        return 'orange'
    elif code in [6, 7, 8]:
        return 'green'
    else:
        return 'gray'

result['Percentage_num'] = result['Percentage'].str.rstrip('%').astype(float)
colors = [get_color(code) for code in result['Education Level Code']]

plt.figure(figsize=(10, 6))

x_labels = result['Education Level Code'].astype(str)
bars = plt.bar(x_labels, result['Percentage_num'], color=colors)

plt.xlabel('Education Level Code')
plt.ylabel('Percentage of Births (%)')
plt.title('Percentage of Births by Education Level Code')

for bar, pct in zip(bars, result['Percentage']):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.1, pct,
             ha='center', va='bottom', fontsize=10)


legend_handles = [
    mpatches.Patch(color='red', label='Low Education (1, 2, 3)'),
    mpatches.Patch(color='orange', label='Medium Education (4, 5)'),
    mpatches.Patch(color='green', label='High Education (6, 7, 8)'),
    mpatches.Patch(color='gray', label='Unknown/Other(-9)')
]
plt.legend(handles=legend_handles, title='Education Level Categories', loc='upper right')

plt.tight_layout()
plt.show()
```

```python
#Sum percentage of births by education group
def assign_group(code):
    if code in [1, 2, 3]:
        return 'Low Education'
    elif code in [4, 5]:
        return 'Medium Education'
    elif code in [6, 7, 8]:
        return 'High Education'
    else:
        return 'Other'


result['Education Group'] = result['Education Level Code'].apply(assign_group)

result['Percentage_num'] = result['Percentage'].str.rstrip('%').astype(float)

percentage_sum_by_group = result.groupby('Education Group')['Percentage_num'].sum().round(1)

percentage_sum_by_group
```

```python
#Yearly Birth Counts and Their Percentage Distribution
birth_per_year=df.groupby('Year')['Number of Births'].sum()
birth_per_year_percent=(birth_per_year/birth_per_year.sum()*100).round(1).astype(str)+'%'
result=pd.concat([birth_per_year,birth_per_year_percent],axis=1)
result.columns=['Number of Births','Percentage']
result
```

```python
# Yearly Birth Distribution by Gender with Percentages
births_by_year_gender = df.groupby(['Year', 'Gender'])['Number of Births'].sum().unstack()
births_by_year_gender = births_by_year_gender.sort_index()
total_births_per_year = births_by_year_gender.sum(axis=1)
births_by_year_gender_percent = (births_by_year_gender.div(total_births_per_year, axis=0) * 100).round(1).astype(str) + '%'
result = pd.concat([births_by_year_gender, births_by_year_gender_percent], axis=1)
result.columns = ['Female', 'Male', 'Female %', 'Male %']
result = result[['Female', 'Female %', 'Male', 'Male %']]

result
```

```python
#Percentage of Births by Gender per Year
import matplotlib.pyplot as plt

result['Female %'] = result['Female %'].str.rstrip('%').astype(float)
result['Male %'] = result['Male %'].str.rstrip('%').astype(float)

plt.figure(figsize=(8,4))
plt.plot(result.index, result['Female %'], marker='o', label='Female %')
plt.plot(result.index, result['Male %'], marker='o', label='Male %')

plt.title('Percentage of Births by Gender per Year')
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.ylim(45, 55)
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Pivot Table of Births by Age Group and Education Level
edu_order = [
    'Unknown or Not Stated',
    '8th grade or less',
    '9th through 12th grade with no diploma',
    'High school graduate or GED completed',
    'Some college credit, but not a degree',
    'Associate degree (AA, AS)',
    "Bachelor's degree (BA, AB, BS)",
    "Master's degree (MA, MS, MEng, MEd, MSW, MBA)",
    'Doctorate (PhD, EdD) or Professional Degree (MD, DDS, DVM, LLB, JD)'
]

df['Education Level'] = df['Education Level Code'].map(edu_map)

pivot_table = df.pivot_table(
    index='Age Group',
    columns='Education Level',
    values='Number of Births',
    aggfunc='sum',
    fill_value=0
)

pivot_table.columns = pd.CategoricalIndex(pivot_table.columns, categories=edu_order, ordered=True)
pivot_table = pivot_table.sort_index(axis=1)

pivot_table
```

```python
#Heatmap of Percentage of Births by Age Group and Education Level Code
births = df.groupby(['Age Group', 'Education Level Code'])['Number of Births'].sum().reset_index()
births_pivot = births.pivot(index='Age Group', columns='Education Level Code', values='Number of Births').fillna(0)
births_pct = births_pivot.div(births_pivot.sum(axis=1), axis=0) * 100
plt.figure(figsize=(12, 7))
sns.heatmap(births_pct, annot=True, fmt=".1f", cmap='coolwarm', cbar_kws={'label': 'Percentage of Births (%)'})
plt.title('Percentage of Births by Age Group and Education Level Code')
plt.ylabel('Age Group')
plt.xlabel('Education Level Code')
plt.tight_layout()
plt.show()
```

```python
#Percentage of Births by Education Level Code Over Time
births_by_edu_year = df.groupby(['Year', 'Education Level Code'])['Number of Births'].sum().reset_index()
total_births_per_year = births_by_edu_year.groupby('Year')['Number of Births'].transform('sum')
births_by_edu_year['Percentage'] = (births_by_edu_year['Number of Births'] / total_births_per_year) * 100

pivot = births_by_edu_year.pivot(index='Year', columns='Education Level Code', values='Percentage')
pivot.plot(marker='o', figsize=(12, 6))
plt.title('Percentage of Births by Education Level Code Over Time')
plt.ylabel('Percentage of Births (%)')
plt.xlabel('Year')
plt.legend(title='Education Level Code', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Impact of COVID-19 on U.S. Births (2016–2021)
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import matplotlib.patches as mpatches


births_by_year = df.groupby('Year', as_index=False)['Number of Births'].sum()


total_births = births_by_year['Number of Births'].sum()
births_by_year['Births %'] = (births_by_year['Number of Births'] / total_births) * 100


colors = ['lightgray' if year < 2020 else 'salmon' for year in births_by_year['Year']]


plt.figure(figsize=(8, 5))
plt.bar(births_by_year['Year'], births_by_year['Births %'], color=colors)

plt.title('Impact of COVID-19 on U.S. Births (2016–2021)')
plt.xlabel('Year')
plt.ylabel('Percentage of Total Births')
plt.xticks(births_by_year['Year'])
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{y:.0f}%'))

legend_handles = [
    mpatches.Patch(color='lightgray', label='Pre-COVID (2016–2019)'),
    mpatches.Patch(color='salmon', label='COVID Impact (2020–2021)')
]
plt.legend(handles=legend_handles, title='Period')

plt.tight_layout()
plt.show()
```

```python
#Percentage of Births by Education Level Code: Pre vs. During/After COVID
df['Period'] = df['Year'].apply(lambda x: 'Pre-COVID' if x < 2020 else 'During/After COVID')
edu_covid = df.groupby(['Education Level Code', 'Period'])['Number of Births'].sum().unstack().fillna(0)
edu_covid_percent = edu_covid.div(edu_covid.sum(axis=1), axis=0) * 100
edu_covid_percent.plot(kind='bar', figsize=(10,6))
plt.title('Percentage of Births by Education Level Code: Pre vs. During/After COVID')
plt.ylabel('Percentage of Births (%)')
plt.xlabel('Education Level Code')
plt.xticks(rotation=0)
plt.legend(title='Period')
plt.tight_layout()
plt.show()
```

```python
# Percentage of Births by State: Pre vs. During/After COVID (Top 10 States)
df['Period'] = df['Year'].apply(lambda x: 'Pre-COVID' if x < 2020 else 'During/After COVID')

state_covid = df.groupby(['State', 'Period'])['Number of Births'].sum().unstack().fillna(0)

state_totals = state_covid.sum(axis=1)

state_covid_percent = state_covid.div(state_totals, axis=0) * 100

top_states = state_totals.sort_values(ascending=False).head(10).index
state_covid_percent.loc[top_states].plot(kind='bar', figsize=(12,7))
plt.title('Percentage of Births by State: Pre vs. During/After COVID (Top 10 States)')
plt.ylabel('Percentage of Births (%)')
plt.xlabel('State')
plt.xticks(rotation=45)
plt.legend(title='Period')
plt.tight_layout()
plt.show()
```

#Conclusion:
Between 2016 and 2021, 48.9% of newborns in the United States were female and 51.1% were male, with males having a slightly higher average birth weight (3307 grams) than females (3195 grams). The data show a slight positive trend between maternal age (24 to 36 years) and average birth weight, although other factors also play a significant role.

The highest proportion of births occurred among women aged 27–28 (28.04%), followed by those aged 31–32 (21.81%) and 25–26 (18.27%), indicating that late twenties to early thirties is the most common childbearing age range in the country.

Maternal age varies by education level: women with doctoral or professional degrees give birth at an average age of 34, those with master’s degrees at 33, while mothers with only a high school diploma or no diploma tend to give birth younger, around 25–26 years. Births are most frequent among women with a high school diploma (25.5%) and bachelor’s degree (20.6%). Overall, 67.1% of births were to mothers without higher education, highlighting a significant educational disparity in birth rates.

From 2016 to 2020, total births in the U.S. declined steadily from approximately 3.95 million to 3.61 million, with a slight increase to 3.66 million in 2021, possibly reflecting recovery following the COVID-19 pandemic.

Regional differences in maternal education are notable. California, one of the states with the highest number of births, shows a diverse range of maternal education levels, predominantly among mothers with high school diplomas and bachelor’s degrees, reflecting its large and socioeconomically varied population. In contrast, states like New York and Illinois have higher proportions of births among mothers with bachelor’s degrees or higher, while Texas and Florida report more births among mothers with less than a high school diploma.

States with fewer births, such as Vermont, North Dakota, and the District of Columbia, often have higher shares of college-educated mothers, indicating demographic and socioeconomic variations.

Overall, U.S. birth trends from 2016 to 2021 highlight the influence of maternal age, education, geography, and external factors like the pandemic on birth outcomes. These insights are crucial for informing public health policies and targeted programs aimed at supporting families and improving maternal and child health.

- Regional differences in birth rates and maternal education should be studied further to ensure resources and services are allocated effectively and equitably.

- The COVID-19 pandemic has impacted birth trends. It’s important to continue monitoring these changes and adjust family planning support as needed.

- Improving data quality and consistency across all states will enable better analysis and more effective policy planning.
